Como a disparidade salarial entre atletas de elite e o trabalhador m√©dio reflete a desigualdade econ√¥mica global, e como isso pode ser transformado em uma oportunidade inovadora no mercado esportivo?

```python
# üì¶ Importando bibliotecas
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# üì• Carregando os dados
jogadores_df = pd.read_excel('/content/jogadores_futebol_basquete.xlsx')
economia_df = pd.read_excel('/content/dados_economicos_python.xlsx')

display(jogadores_df)
display(economia_df)
```

## üìä Estat√≠sticas Descritivas

```python
# M√©dia, Mediana, Desvio Padr√£o por esporte
jogadores_df.groupby('Esporte')['Sal√°rio Anual (US$ milh√µes)'].agg(['mean', 'median', 'std'])
```

```python
# Correla√ß√£o entre PIB e Sal√°rio M√©dio
economia_df[['PIB (US$ trilh√µes/bilh√µes)', 'Sal√°rio M√©dio Anual (US$)', 'Infla√ß√£o (%)']].corr()
```

```python
# Correla√ß√£o entre IDH e outros indicadores econ√¥micos
correlacoes = economia_df[["IDH", "PIB (US$ trilh√µes/bilh√µes)", "Sal√°rio M√©dio Anual (US$)", "Infla√ß√£o (%)"]].corr()
correlacoes
```

# Vari√°veis Comparadas;
###	Correla√ß√£o com IDH;
    Correla√ß√£o moderada positiva,
    pa√≠ses com maior PIB tendem a ter maior IDH.

###  Sal√°rio M√©dio Anual;
	Correla√ß√£o forte positiva,
    quanto maior o sal√°rio m√©dio, maior o IDH.

####  Infla√ß√£o (%);
	Correla√ß√£o forte negativa,
    quanto maior a infla√ß√£o, menor tende ser o IDH.

    
Esses resultados mostram que economias com sal√°rios mais altos e infla√ß√£o mais controlada tendem a ter melhor qualidade de vida (IDH mais alto).
    

## üìà Visualiza√ß√µes Interativas
